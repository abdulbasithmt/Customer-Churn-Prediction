```python
import pandas as pd
```

```python
df=pd.read_csv('/content/churn.csv')
df
```

```python
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
df.drop('customerID', axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X=df.drop('Churn_Yes', axis=1)
y=df['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

1. The model achieved approximately 80% accuracy.
2. Precision and recall metrics indicate balanced performance.
3. The model can be used to identify customers likely to churn, helping businesses take preventive actions.
